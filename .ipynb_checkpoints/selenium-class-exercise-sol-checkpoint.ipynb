### What is Selenium?
---

Selenium is a web automation framework that can be used to automate website testing. Because Selenium starts a webbrowser, it can do any task you would normally do on the web (ex: It allows us to render JavaScript just as a human-navigated browser would)

## Let's Scrape (OpenTable) Website

<br>

[OpenTable](https://www.opentable.com/washington-dc-restaurant-listings) is an online table booking in restaurants. Let's scrape their site to get how many times each restaurant has booked.


let's start by explore the structure of the page

```python
import re
import requests
import pandas as pd
from bs4 import BeautifulSoup
from time import sleep
from selenium import webdriver
```

### 1- Requests

```python
## Step 1 : Download Html page 
```

```python
response = requests.get('https://www.opentable.com/washington-dc-restaurant-listings?currentview=list&size=100&sort=Popularity&covers=2&dateTime=2019-04-02+19%3A00&latitude=38.902777&longitude=-77.12594&metroId=9')
```

### 2- Selenium


```python
## Step 1 : Download Html page 
```

```python
# open browser
driver = webdriver.Chrome(executable_path='./chromedriver/chromedriver')

# request url
driver.get('https://www.opentable.com/washington-dc-restaurant-listings?currentview=list&size=100&sort=Popularity&covers=2&dateTime=2019-04-02+19%3A00&latitude=38.902777&longitude=-77.12594&metroId=9')

#wait
#sleep(3)
driver.implicitly_wait(2)

#catch data
HTML = driver.page_source

# close  browser
driver.close()
```

```python
# step 2 : parsing
```

```python
<div class="booking"><span class="tadpole"></span>Booked 90 times today</div>
```

```python
soup = BeautifulSoup(HTML, 'lxml')
```

```python
item_lst = soup.find_all('div', attrs= {'class':'booking'})
```

```python
for i in item_lst:
    print(i.text)
```

##  Web Automation With Selenium
<br>
Let's do automated Search for `apple` phones in [(Souq Saudi)](https://saudi.souq.com/sa-en/apple/mobile-phones-33/new/a-t-c/s/?_=1459842493768&page=1&ref=nav)

```python
# open browser
driver = webdriver.Chrome(executable_path='./chromedriver/chromedriver')

# request url
driver.get('https://saudi.souq.com/sa-en/')

# search box
elem = driver.find_element_by_id('search_value')

# clear search box
elem.clear()

# send my search words
elem.send_keys("apple")


# click button
button = driver.find_element_by_id('searchButton')
# press click 
button.click()

# save page
HTML = driver.page_source
#close browser
driver.close()
```
